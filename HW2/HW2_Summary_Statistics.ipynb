# **Measure the Central Tendency of the Philly Park Sizes**

Measure the mean and median and compare them.

```python
# Show the mean size of the dataframe and print it
meanps = round(ppark['ACREAGE'].mean(), 4)
print("The Mean is:", meanps)
```

```python
# Show the median size of the dataframe and print it
medianps = round(ppark['ACREAGE'].median(), 4)
print("The Median is:", medianps)
```

**1) What can you conclude from the values of the mean and median you extracted?**


> The mean and median are not the same and have a moderate difference in values (mean = ~19.95 acres and median = ~3.60 acres). Although we do not know the mode, we can make the assumption that since the mean and median do not equal each other, this data **does not** have a central tendency (normally distributed curve). This may indicate that the data is skewed.

**2) What can you say about the distribution of the ACREAGE variable in the data?**


> By only examining the value outputs, we know that since we lack central tendency, there will be more variance. This data is likely skewed to the *right* due to the mean being larger than the median. By examining the graph, we can actually see the distribution of park sizes, with a few outliers above 500 acres. This display confirms that the data is skewed to the *right*.





```python
# Visualize a histogram of the park size
ax = ppark['ACREAGE'].hist(bins=15)
ax.set_xlabel('Philly Park Size (acres)')
ax.set_ylabel('Number of Parks')
ax.set_title('Figure 1. Distribution of Park Sizes in Philadelphia, USA')
ax.axvline(meanps, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(medianps, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([meanps, medianps], labels = ['Mean', 'Median']);
```

**3) Describe what the graph means.**


> From the graph, we can see that the mean is greater than the median, which means that the data is skewed to the right. This essentially tells us that the ACREAGE of most parks fall on the left side of the graph, indicating that the park sizes are typically smaller, less than about 200 acres or so, for most of the parks in Philadelphia.


# **4) What is the spread of park sizes?**

Calculate the standard deviation of the park sizes.


> A standard deviation of around 112 acres means that most park sizes fall within 112 acres of the mean (~19.95 acres). There is definitely some spread within the data as some park sizes are greater than 1000 acres.



```python
# Calculating the standard deviation of the park sizes
stdps = round(ppark['ACREAGE'].std(), 4)
print("The standard deviation of the park sizes is", stdps)
```

## **5) Select only a subset of the parks: the parks smaller than 10 acres.**

```python
# Removing Unknown
ax = ppark[~(ppark["ACREAGE"]=="Unknown")]["ACREAGE"].unique()
smallparks = ppark.loc[ppark['ACREAGE'] < 10]
smallparks.head()
```

```python
# Calculate the mean and median of the parks with size less than 10 acres
meansp = round(smallparks['ACREAGE'].mean(), 4)
print("The Mean is:", meansp)

mediansp = round(smallparks['ACREAGE'].median(), 4)
print("The Median is:", mediansp)
```

```python
# Plot a histogram of the parks with size less than 10 acres with the mean and median indicated on the plot
ax = smallparks['ACREAGE'].hist(bins=25)
ax.set_xlabel('Park Size (acres)')
ax.set_ylabel('Number of Parks')
ax.set_title('Figure 2. Distribution of Park Sizes less than 10 Acres in Philadelphia, USA')
ax.axvline(meansp, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(mediansp, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([meansp, mediansp], labels = ['Mean', 'Median']);
```

```python
# Grouping the parks based on the parent name
group_park = ppark.groupby(['PARENT_NAME']).sum()
group_park
```

**7) Using sns.pairplot plot the pairplot of ACREAGE and Shape__Area variables. What is the relationship that you can see?**


> There is a positive correlation between ACREAGE and Shape__Area. This means that as the ACREAGE increases, so does the Shape_Area.



```python
# Plotting pairplot
pair = ['ACREAGE', 'Shape__Area']
pair
```

```python
sns.pairplot(group_park[pair], size=2.0)
```

```python
# Calculate the mean and median of parks grouped by parent name
groupmean = round(group_park['ACREAGE'].mean(), 4)
groupmedian = round(group_park['ACREAGE'].median(), 4)
print('The Mean of the Grouped Parks is:', groupmean)
print('The Median of the Grouped Parks is:', groupmedian)
```

**8) Visualize the distribution of the parks grouped by parent name.**

```python
# Visualizing ditribtuion of parks grouped by parent name
ax = group_park['ACREAGE'].hist(bins=15)
ax.axvline(groupmean, color='red', linestyle='solid', linewidth=1.5, alpha = .8)
ax.axvline(groupmedian, color='green', linestyle='solid', linewidth=1.5, alpha = .8)
ax.legend([groupmean, groupmedian], labels = ['Mean', 'Median'])
ax.set_xlabel('Park Size (acres)')
ax.set_ylabel('Number of Parks')
pylab.title('Figure 3. Distribution of Parks Grouped by Parent Name in Philadelphia, USA')
```

**9) Explain what the graph means.**


> This graph, like previous graphs, is skewed to the right, with outliers reaching above 500, 1000, 1500, and 2000 acres. For the distribution of parks that are grouped by parent name, the **majority parks** are on the **smaller side** still with sizes of less than 250 acres.
