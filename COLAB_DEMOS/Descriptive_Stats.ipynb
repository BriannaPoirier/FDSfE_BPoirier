```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/kirenz/datasets/master/wage.csv")
data
```

```python
# The first 5 rows
data.head()
```

```python
# The Last 5 rows
data.tail()
```

```python
data.shape
```

```python
# Summary of the stats of the numerical variables
data.describe()
```

```python
# Accessing specific variables
wage = data['wage']
wage
```

```python
# Calculating the mean wage
mean_wage = round(wage.mean(), 4)
print('The mean for the variable wage in the data set is', mean_wage)
```

```python
# Accessing specific variables
age = data['age']
age
```

```python
# Calculating the mean age
mean_age = round(age.mean(), 4)
print('The mean for the variable age in the data set is', mean_age)
```

```python
mean_age = round(data['age'].mean(), 4)
print(mean_age)
```

```python
median_age = round(data['age'].median(), 4)
print(median_age)
```

```python
# Accessing specific variables
logwage = data['logwage']
logwage
```

```python
# Calculating the mean logwage
mean_logwage = round(logwage.mean(), 4)
print('The mean for the variable logwage in the data set is', mean_logwage)
```

```python
mean_logwage = round(data['logwage'].mean(), 4)
print(mean_logwage)
```

```python
print('mean logwage: ', round(data['logwage'].mean(), 4))
```

```python
# Max, Min, Range

print('Max age is ', data['age'].max())
print('Min age is ', data['age'].min())
print('Range age is ', data['age'].max() - data['age'].min())
print('Median age is ', data['age'].median())
```

```python
# Stats summary by variables
x = data['age'].groupby(data['education'])
x.describe()
```
